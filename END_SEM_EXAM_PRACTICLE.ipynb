#Question a
Write a Python program to find the first
triangle number to have over n(given) divisors.

From Wikipedia: A triangular number is a number that is the sum of all
of the natural numbers up to a certain number. For example, 10 is a triangular
number because 1 + 2 + 3 + 4 = 10. The first 25 triangular numbers are: 1, 3,
6, 10, 15, 21, 28, 36, 45, 55, 78, 91, 105, 120, 136, 153, 171, 190, 210, 231,
253, 276, 300, 325, and 351.

A triangular number is calculated by the equation: n(n+1)/2

The factors of the first five triangle numbers:<br>

1: 1<br>
3: 1, 3<br>
6: 1, 2, 3, 6<br>
10: 1, 2, 5, 10<br>
15: 1, 3, 5, 15<br>

In the above list 6 is the first triangle number to have over four
divisors.


```python
# Function to retrieve all divisors of a number
def get_divisors(num):
    divisors = set()
    i = 1
    # Iterate up to the square root for efficiency
    while i * i <= num:
        if num % i == 0:
            divisors.add(i)
            divisors.add(num // i)
        i += 1
    return sorted(divisors)

# Function to find the triangular number exceeding n divisors
def first_triangle_over_n_divisors(n):
    k = 1
    while True:
        triangle = k * (k + 1) // 2
        divisors = get_divisors(triangle)
        if len(divisors) > n:
            return k, triangle, divisors
        k += 1

# User input
n = 4  # fixed for demonstration, can be replaced with input()

# Compute result
k, triangle, divisors = first_triangle_over_n_divisors(n)

# Output
print("---------- TASK 1 OUTPUT ----------")
print(f"Given n: {n}")
print(f"Triangular Index     : {k}")
print(f"Triangular Number    : {triangle}")
print(f"Number of Divisors   : {len(divisors)}")
print(f"Divisors             : {divisors}")
print("-----------------------------------")
```

#### **Explanation**
A triangular number is defined by the formula Tₙ = n(n + 1) / 2.  
The program iteratively generates each triangular number and determines its divisors.

Key points:
- Divisors are computed efficiently by checking only up to √n.
- Each triangular number is tested until one is found with more than the required number of divisors.
- The program outputs the triangular index, the value, and its complete divisor list.

#Question b
Write a Python program to find the starting
number, under ten thousand will create the longest chain.


From Wikipedia - The Collatz conjecture is a conjecture in mathematics
that concerns a sequence defined as follows: start with any positive integer n.
Then each term is obtained from the previous term as follows: if the previous
term is even, the next term is one half the previous term. If the previous term
is odd, the next term is 3 times the previous term plus 1. The conjecture is
that no matter what value of n, the sequence will always reach 1.


Python: Find the maximum total from top to bottom of the triangle below.

```python
# Function generating the full Collatz sequence
def collatz_sequence(n):
    seq = [n]
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3*n + 1
        seq.append(n)
    return seq

# Function computing the longest chain under a given limit
def longest_collatz_under(limit):
    max_len = 0
    best_start = 1
    best_chain = []

    for num in range(1, limit):
        chain = collatz_sequence(num)
        if len(chain) > max_len:
            max_len = len(chain)
            best_start = num
            best_chain = chain

    return best_start, max_len, best_chain

# Execute task
limit = 10000
start, length, chain = longest_collatz_under(limit)

# Output
print("---------- TASK 2 OUTPUT ----------")
print(f"Limit                 : {limit}")
print(f"Starting Number       : {start}")
print(f"Chain Length          : {length}")
print(f"Collatz Chain         : {chain}")
print("-----------------------------------")
```

#### **Explanation**

The Collatz sequence is generated using:
- n → n/2 when n is even  
- n → 3n + 1 when n is odd  

This program evaluates all starting values below 10,000 and tracks:
- The starting number
- The total chain length
- The full resulting sequence

The longest chain encountered is reported in structured output.

#Question c
Write a program to display the cos(x) and
tan(x) value where x ranges from 0 to 360 in steps of 15.

```python
import math

def display_trigonometric_values():
    """
    Calculates and displays the values of cos(x) and tan(x)
    for x from 0 to 360 degrees in steps of 15 degrees.
    """
    # Define the range and step for degrees
    start_degree = 0
    end_degree = 360
    step = 15

    # --- Print Header using f-string formatting ---
    # Widths: Degree (10), Radians (15), Cos(x) (15), Tan(x) (15)
    print(f"{'---':=^55}")
    print(f"{'TRIGONOMETRIC VALUES (0° to 360°)' :^55}")
    print(f"{'---':=^55}")
    print(f"{'Degrees (x)':<10}{'Radians':<15}{'cos(x)':>15}{'tan(x)':>15}")
    print(f"{'-'*55}")

    # Iterate through the degree values
    for x_deg in range(start_degree, end_degree + 1, step):

        # 1. Convert degrees to radians
        # Formula: radians = degrees * (pi / 180)
        x_rad = math.radians(x_deg)

        # 2. Calculate cos(x)
        cos_x = math.cos(x_rad)

        # 3. Calculate tan(x)
        # Note: tan(x) is undefined at 90 and 270 degrees.
        # We check the degree to prevent a large, unusable number.
        if x_deg % 180 == 90: # Check for 90, 270, etc.
            tan_x_str = "UNDEFINED"
        else:
            tan_x = math.tan(x_rad)
            # Format tan(x) to 4 decimal places
            tan_x_str = f"{tan_x:>15.4f}"


        # 4. Display the results using f-string formatting
        # Formatting:
        print(f"{x_deg:<10}{x_rad:<15.4f}{cos_x:>15.4f}{tan_x_str}")

    print(f"{'':=^55}")

# Execute the function
display_trigonometric_values()
```

#### **Explanation**
- The program iterates through angles from 0° to 360° in steps of 15°.
- Each degree value is converted to radians using the `math.radians()` function.
- The cosine of the angle is computed using `math.cos()` and formatted to four decimal places.
- The tangent is calculated using `math.tan()`, except at 90° and 270°, where it is marked as UNDEFINED.
- A structured table is printed using formatted f-strings to align degree, radian, cosine, and tangent values.
- The output section is visually separated with headers and dividers for readability.
