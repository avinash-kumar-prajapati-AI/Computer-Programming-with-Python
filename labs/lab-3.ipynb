#### 1.2 If else statement

```python
# Q4. WAP whether a person is eligible for vote or not, if not then how many year are left to be eligible for vote.
age = 10
if age > 18:
    print("Eligible for vote")
else:
    print(18-10, "Years are left to be eligible for vote.")

# using ternary operator
(
    print("Eligible for vote")
    if age > 18
    else print(18 - 10, "Years are left to be eligible for vote.")
)
```
