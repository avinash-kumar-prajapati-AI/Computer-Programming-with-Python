```python
# Q2. Write a program to calculate area of a triangle using hero's formula
edges = input("Enter the three side lenght of the Triangle. like this a,b,c").split(",")
print()
a, b, c = int(edges[0]), int(edges[1]), int(edges[2])
s = a+b+c
area = math.sqrt(s*(s-a)*(s-b)*(s-c))
print("Area of this traingle: ", area)
```

```python
# Q5 Write a program to calculate student's result based on two examinations, 1 sport event and 3 activities. Conductedweightage of activities 30%, sports 20% and examination = 50%

activity_weight = 30/3
examination_weight  = 50
sport_weight = 20

score_in_sport = input("Enter th sport score out of 100: ")

score_in_activity_1 = input("Enter the score in activity 1 out of 100 : ")
score_in_activity_2 = input("Enter the score in activity 2 out of 100 : ")
score_in_activity_3 = input("Enter the score in activity 3 out of 100 : ")

score_in_examination = input("Enter the score in examination out of 100: ")

sport = (int(score_in_sport)*sport_weight)/100
activity1 = (int(score_in_activity_1) * activity_weight) / 100
activity2 = (int(score_in_activity_2) * activity_weight) / 100
activity3 = (int(score_in_activity_3) * activity_weight) / 100

exam = (int(score_in_examination)*examination_weight)/100

result = sport + activity1 + activity2 +activity3 +exam
print("Total score in percentage in sport : ", sport,"%", " out of 20%")
print("Total score in percentage in activities : ", activity1+activity2+activity3, "%", "out of 30%.")
print("Total score in percentage in examination : ", exam, "%", "out of 50%.")
print(format("", "->50"))
print("Your Result based on the respective scores provided : ", result)
```
