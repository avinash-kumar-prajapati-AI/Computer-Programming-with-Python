1. Write a Python program to create and display all combinations of letters, selecting each
letter from a different key in a dictionary.
<br>
<br>
Sample data: {&#39;1&#39;:[&#39;a&#39;,&#39;b&#39;], &#39;2&#39;:[&#39;c&#39;,&#39;d&#39;]}
<br>
<br>
Expected Output:
- ac
- ad
- bc
- bd

```python
from itertools import product

# Solution 1
combinations = lambda data: ["".join(comb) for comb in product(*data.values())]

# Test
data = {"1": ["a", "b"], "2": ["c", "d"]}
print(*combinations(data), sep="\n")
```

2. Write a Python program to get the top three items in a shop.
<br>
<br>
Sample data: {&#39;item1&#39;: 45.50, &#39;item2&#39;:35, &#39;item3&#39;: 41.30, &#39;item4&#39;:55, &#39;item5&#39;: 24}
<br>
<br>
Expected Output:
- item4 55
- item1 45.5
- item3 41.3

```python
top_three = lambda items: [
    (k, v) for k, v in sorted(items.items(), key=lambda x: x[1], reverse=True)[:3]
]

# Test
shop = {"item1": 45.50, "item2": 35, "item3": 41.30, "item4": 55, "item5": 24}
for item, price in top_three(shop):
    print(item, price)
```

3. Write a Python program to find the specified number of maximum values in a given dictionary.
<br>
<br>
Original Dictionary:
{&#39;a&#39;: 5, &#39;b&#39;: 14, &#39;c&#39;: 32, &#39;d&#39;: 35, &#39;e&#39;: 24, &#39;f&#39;: 100, &#39;g&#39;: 57, &#39;h&#39;: 8, &#39;i&#39;: 100}
<br>
<br>
1 maximum value(s) in the said dictionary:
[&#39;f&#39;]
<br>
2 maximum value(s) in the said dictionary:
[&#39;f&#39;, &#39;i&#39;]
<br>
5 maximum value(s) in the said dictionary:
[&#39;f&#39;, &#39;i&#39;, &#39;g&#39;, &#39;d&#39;, &#39;c&#39;]

```python
max_values = lambda d, n: [
    k for k, _ in sorted(d.items(), key=lambda x: x[1], reverse=True)[:n]
]

original_dict = {
    "a": 5,
    "b": 14,
    "c": 32,
    "d": 35,
    "e": 24,
    "f": 100,
    "g": 57,
    "h": 8,
    "i": 100,
}
print(f"1 maximum value(s): {max_values(original_dict, 1)}")
print(f"2 maximum value(s): {max_values(original_dict, 2)}")
print(f"5 maximum value(s): {max_values(original_dict, 5)}")
```

4. Write a Python program to combine two lists into a dictionary. The elements of the first one serve as keys and the elements of the second one serve as values. Each item in the first list must be unique and hashable.
<br>
<br>
Original lists:
<br>
[&#39;a&#39;, &#39;b&#39;, &#39;c&#39;, &#39;d&#39;, &#39;e&#39;, &#39;f&#39;]
<br>
[1, 2, 3, 4, 5]
<br>
Combine the values of the said two lists into a dictionary:
{&#39;a&#39;: 1, &#39;b&#39;: 2, &#39;c&#39;: 3, &#39;d&#39;: 4, &#39;e&#39;: 5}

```python
combine_lists = lambda keys, values: {k: v for k, v in zip(keys, values)}

keys = ["a", "b", "c", "d", "e", "f"]
values = [1, 2, 3, 4, 5]
print(combine_lists(keys, values))
```

5. Write a Python function that evaluates mathematical expressions represented as dictionaries. The expression dictionary should support basic operations (+, -, *, /), variables, and nested expressions.

```python
evaluate = lambda expr, vars_dict={}: (
    expr
    if isinstance(expr, (int, float))
    else (
        vars_dict.get(expr, expr)
        if isinstance(expr, str)
        else {
            "+": lambda: evaluate(expr["left"], vars_dict)
            + evaluate(expr["right"], vars_dict),
            "-": lambda: evaluate(expr["left"], vars_dict)
            - evaluate(expr["right"], vars_dict),
            "*": lambda: evaluate(expr["left"], vars_dict)
            * evaluate(expr["right"], vars_dict),
            "/": lambda: evaluate(expr["left"], vars_dict)
            / evaluate(expr["right"], vars_dict),
        }[expr["op"]]()
    )
)

expression1 = {"op": "+", "left": 5, "right": 3}
expression2 = {"op": "*", "left": {"op": "+", "left": 2, "right": 3}, "right": 4}
expression3 = {"op": "+", "left": "x", "right": "y"}

print(evaluate(expression1))  # 8
print(evaluate(expression2))  # 20
print(evaluate(expression3, {"x": 10, "y": 5}))  # 15
```
